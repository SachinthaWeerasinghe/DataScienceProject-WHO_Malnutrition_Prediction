## **BOXPLOT FOR THE AGE GROUPS OF THE CHILDREN TO IDENTIFY THE OUTLIERS**

```python
# Inspect unique values in the "AGEGROUP" column
unique_values = df['AGEGROUP'].unique()
print("Unique values in 'AGEGROUP' column:")
print(unique_values)

# Function to calculate the midpoint of a range
def calculate_midpoint(age_range):
    if pd.isna(age_range):
        return np.nan
    # Convert the value to string if it's not already
    age_range = str(age_range)
    # Split the range string into two parts
    parts = age_range.split('-')
    if len(parts) != 2:
        print(f"Unexpected format: {age_range}")
        return np.nan
    # Convert the parts to float and calculate the midpoint
    start = float(parts[0].strip())
    end = float(parts[1].strip())
    midpoint = (start + end) / 2
    return midpoint

# Ensure all values in the "AGEGROUP" column are strings
df['AGEGROUP'] = df['AGEGROUP'].astype(str)

# Apply the function to the "AGEGROUP" column to get midpoints
df['AGE_MIDPOINT'] = df['AGEGROUP'].apply(calculate_midpoint)

# Print the modified DataFrame
#print("Modified DataFrame with Age Midpoints:")
#print(df)

# Create a box plot for the age midpoints
plt.figure(figsize=(10, 6))
plt.boxplot(df['AGE_MIDPOINT'].dropna(), vert=False)  # Drop NaN values for plotting

# Set plot labels and title
plt.xlabel('Age Midpoint')
plt.title('Box Plot of Ages of the Selected Children')

# Display the plot
plt.show()
```

```python
# Function to calculate the midpoint of a range
def calculate_midpoint(age_range):
    if pd.isna(age_range):
        return np.nan
    # Convert the value to string if it's not already
    age_range = str(age_range)
    # Split the range string into two parts
    parts = age_range.split('-')
    if len(parts) != 2:
        print(f"Unexpected format: {age_range}")
        return np.nan
    # Convert the parts to float and calculate the midpoint
    start = float(parts[0].strip())
    end = float(parts[1].strip())
    midpoint = (start + end) / 2
    return midpoint

# Ensure all values in the "AGEGROUP" column are strings
df['AGEGROUP'] = df['AGEGROUP'].astype(str)

# Apply the function to the "AGEGROUP" column to get midpoints
df['AGE_MIDPOINT'] = df['AGEGROUP'].apply(calculate_midpoint)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['AGE_MIDPOINT'].quantile(0.25)
Q3 = df['AGE_MIDPOINT'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Determine the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['AGE_MIDPOINT'] < lower_bound) | (df['AGE_MIDPOINT'] > upper_bound)]

# Count the number of outliers
num_outliers = outliers.shape[0]

# Calculate the total number of rows
total_rows = df.shape[0]

# Calculate the percentage of outliers
percentage_outliers = (num_outliers / total_rows) * 100

print("Number of outliers in 'AGE_MIDPOINT' column:", num_outliers)
#print("Total number of rows:", total_rows)
print(f"Percentage of outliers: {percentage_outliers:.2f}%")


# Optional: print the outliers
#print("Outliers:")
#print(outliers)
```
